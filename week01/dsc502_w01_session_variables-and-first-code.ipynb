## Part 2 — Data types

Python needs to know what *kind* of value a variable holds — its **data type**. The four most important right now:

| Type | What it holds | Example |

The `type()` function tells you which type a variable is. Run the example below.

````python
# Worked example — type() on one variable
platform_name = "QuickBite"
print(platform_name)
print(type(platform_name))
````

## Part 3 — f-strings

Sooner or later you want to put a variable's value *inside* a sentence. You could try joining things with `+`:

```python
print("Delivery time: " + delivery_time_mins + " minutes")
```

But that fails if `delivery_time_mins` is a number. An **f-string** solves this cleanly. Write `f` before the opening quote, then put variable names in curly braces `{}`:

```python
print(f"Delivery time: {delivery_time_mins} minutes")
```

Python replaces `{delivery_time_mins}` with the actual value. Run the example below.

````python
# Worked example — one variable in an f-string
platform_name = "QuickBite"
print(f"Welcome to {platform_name}!")
````

### Indexing — getting one item out

Python counts from **zero**. The first item in a list is at index `[0]`.

Negative indices count from the end: `[-1]` is always the last item.

````python
# Worked example — indexing
restaurants = ["Burger Palace", "Pizza Roma", "Sushi Garden", "Taco Town"]

print(restaurants[0])    # First item
print(restaurants[1])    # Second item
print(restaurants[-1])   # Last item
````

### `len()`, `.append()`, and slicing

````python
# Worked example — len, append, slice
restaurants = ["Burger Palace", "Pizza Roma", "Sushi Garden", "Taco Town"]

print(len(restaurants))          # How many items?

restaurants.append("The Noodle Bar")  # Add a new item
print(restaurants)               # See the updated list

print(restaurants[0:3])          # First three items (0, 1, 2 — not 3)
````

- **TypeError** — you used the wrong data type for an operation
